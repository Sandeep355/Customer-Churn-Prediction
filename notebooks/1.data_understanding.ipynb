## Data Dictionary

| Column | Data Type | Description |
|---------|----------|-------------|
| customerID | Object | Unique identifier for each customer |
| gender | Object | Customer's gender |
| SeniorCitizen | Integer | 1 = Senior Citizen, 0 = Not Senior Citizen |
| Partner | Object | Whether the customer has a partner |
| Dependents | Object | Whether the customer has dependents |
| tenure | Integer | Number of months the customer has stayed |
| PhoneService | Object | Whether the customer has phone service |
| InternetService | Object | Type of internet service |
| Contract | Object | Contract type |
| PaymentMethod | Object | Customer's payment method |
| MonthlyCharges | Float | Monthly subscription charges |
| TotalCharges | Float/Object | Total amount charged to the customer |
| Churn | Object | Target variable (Yes = Churned, No = Retained) |

```python
## copying the data

df_clean=df.copy()
```

```python
df_clean.isnull().sum()
```

```python
(df_clean == "").sum()
```

```python
(df_clean["Total Charges"] == " ").sum()
```

```python
import numpy as np

df_clean["Total Charges"] = df_clean["Total Charges"].replace(" ", np.nan)
```

```python
df_clean["Total Charges"] = pd.to_numeric(df_clean["Total Charges"])
```
